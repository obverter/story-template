```python
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get('https://www.bbc.com/')
doc = BeautifulSoup(response.text)
```

```python
# Grab all of the titles 
titles = doc.select(".media__title a")
titles
```

```python
for title in titles:
    row = {}
    # title
    row['title'] = title.text.strip()
    # link
    row['href'] = title['href']
```
